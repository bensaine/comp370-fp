```python
import pandas as pd
from dotenv import load_dotenv
load_dotenv()
import os
import requests
from time import sleep
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
articles = pd.read_csv('articles.csv')
sources = pd.read_csv('sources_bias.csv')
sources.set_index('source_id', inplace=True)
articles['bias'] = articles['source_id'].map(sources['bias'])
articles.to_csv('articles_bias.csv', index=False)
```

```python
articles_bias = pd.read_csv('articles_bias.csv')
# pie chart of bias 
bias_counts = articles_bias['bias'].value_counts()
bias_counts.plot.pie(autopct='%1.1f%%')
sns.set_theme()
plt.title('Proportional Distribution of Political Bias from Sources')
plt.ylabel('')
plt.show()
```
